```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x1 = np.linspace(1, 10, 100)
x2 = np.linspace(1, 10, 100)[:, np.newaxis]
x3 = np.exp(-x1) ** 10 + np.cos(x2 * x1 + np.exp(x1**0.5 / x2))
plt.imshow(x3, origin = 'lower', extent = [0, 5, 0, 5], cmap = 'viridis')
plt.colorbar();
```

```python
x1 = np.arange(100)
x2 = np.arange(100)
x3 = np.multiply.outer(x1,x2)
plt.imshow(x3, origin = 'lower', extent = [0, 5, 0, 5], cmap = 'viridis')
plt.colorbar();
```
