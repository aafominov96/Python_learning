```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn
```

```python
#Импорт данных
data = pd.read_csv('president_heights.csv')
heights = np.array(data['height(cm)'])
print(heights)
```

```python
#Вывод некоторых сводных данных
print('Mean height: ', heights.mean())
print('Standart deviation: ', heights.std())
print('Minimum height: ', heights.min())
print('Maximum height: ', heights.max())
print('25th percentile: ', np.percentile(heights, 25))
print('50th percentile: ', np.percentile(heights, 50))
print('75th percentile: ', np.percentile(heights, 75))
print('Median ', np.median(heights))
heights_mean = heights.mean()
heights_diff = heights_mean - heights
```

```python
seaborn.set()
plt.hist(heights)
plt.title('Height Distribution of US Presidents')
plt.xlabel('height(cm)')
plt.ylabel('number');
```

```python
plt.plot(heights_diff);
plt.title('Difference heights and mean height')
plt.xlabel('num of president')
plt.ylabel('difference heights(cm)');
```

```python
seaborn.set()
np.random.seed(0)
indices = np.random.choice(heights.shape[0], int(heights.shape[0] / 2), replace = False)
selection = heights[indices]
x_axis = np.arange(heights.shape[0])
plt.hist(selection);
```

```python
bins = np.linspace(heights.min(), heights.max(), heights.max()-heights.min())
counts = np.zeros_like(bins)
i = np.searchsorted(bins, heights)
np.add.at(counts, i, 1)
plt.plot(bins, counts, linestyle = 'steps');
```
